Obtenci贸n del gradiente morfol贸gico de una imagen

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.morphology import disk, dilation, erosion
```

```python
imagen = io.imread('brain-scan.jpeg', as_gray=True)

# Conversi贸n a formato de 8 bits (0-255)
imagen = (imagen * 255).astype(np.uint8)

selem = disk(2)

# Calculo del gradiente morfol贸gico
gradiente = dilation(imagen, selem) - erosion(imagen, selem)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(imagen, cmap='gray')
ax[0].set_title('Imagen Original')
ax[0].axis('off')

ax[1].imshow(gradiente, cmap='gray')
ax[1].set_title('Gradiente Morfol贸gico')
ax[1].axis('off')

plt.show()
```
