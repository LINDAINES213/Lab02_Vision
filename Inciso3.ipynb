## Inciso 3

Aplicación de distintas operaciones morfológicas a cada una de las imágenes siguientes.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.morphology import disk, dilation, erosion, opening, closing, white_tophat
```

```python
butterfly = io.imread('butterfly.jpeg')
quetzal = io.imread('quetzalgris.png')
chest_xray = io.imread('chestXray.jpeg')
```

```python
# Conversión a escala de grises

def to_grayscale(image):
    if len(image.shape) == 3 and image.shape[2] == 4:  
        image = color.rgba2rgb(image)  
    if len(image.shape) == 3: 
        image = color.rgb2gray(image)
    return image

butterfly = to_grayscale(butterfly)
quetzal = to_grayscale(quetzal)
chest_xray = to_grayscale(chest_xray)
```

```python
selem = disk(5)

def apply_morphological_ops(image, title):
    
    # Dilatación
    dilated = dilation(image, selem)
    
    # Erosión
    eroded = erosion(image, selem)
    
    # Opening
    opened = opening(image, selem)
    
    # Closing
    closed = closing(image, selem)
    
    # White Top-Hat
    white_tophat = white_tophat(image, selem)
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    fig.suptitle(title, fontsize=16)
    
    axes[0, 0].imshow(image, cmap='gray')
    axes[0, 0].set_title('Original')
    axes[0, 0].axis('off')
    
    axes[0, 1].imshow(dilated, cmap='gray')
    axes[0, 1].set_title('Dilatación')
    axes[0, 1].axis('off')
    
    axes[0, 2].imshow(eroded, cmap='gray')
    axes[0, 2].set_title('Erosión')
    axes[0, 2].axis('off')
    
    axes[1, 0].imshow(opened, cmap='gray')
    axes[1, 0].set_title('Opening')
    axes[1, 0].axis('off')
    
    axes[1, 1].imshow(closed, cmap='gray')
    axes[1, 1].set_title('Closing')
    axes[1, 1].axis('off')
    
    axes[1, 2].imshow(white_tophat, cmap='gray')
    axes[1, 2].set_title('White Top-Hat')
    axes[1, 2].axis('off')
    
    plt.show()
```

```python
apply_morphological_ops(butterfly, 'Butterfly')
apply_morphological_ops(quetzal, 'Quetzal')
apply_morphological_ops(chest_xray, 'Chest X-Ray')
```
